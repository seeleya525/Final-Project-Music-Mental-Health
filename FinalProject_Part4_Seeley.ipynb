## Question 1: Does listening to more music in a day correlate to worse mental health?

```python
import matplotlib.pyplot as plt
```

For our x-axis, we'll use the time in hours that a person spends listening each day. On the y-axis, we'll plot the mental health rank (0-10) provided by each respondent for each condition. We'll use a scatter plot to see if we can find a relation between our two numerical variables.

```python
df.plot.scatter(x='hours', y='anxiety', c='red', alpha = 0.1)
plt.title('Anxiety Levels are Not Affected by Time Spent Listening to Music')
plt.xlabel('Hours/Day Spent Listening to Music')
plt.ylabel('Level of Reported Anxiety')
```

```python
df.plot.scatter(x='hours', y='depression', c='blue', alpha = 0.1)
plt.title('Depression Levels are Not Impacted by Time Spent Listening to Music')
plt.xlabel('Hours/Day Spent Listening to Music')
plt.ylabel('Level of Reported Depression')
```

```python
df.plot.scatter(x='hours', y='insomnia', c='green', alpha = 0.1) 
plt.title('Insomnia Severity is Not Impacted by Time Spent Listening to Music')
plt.xlabel('Hours/Day Spent Listening to Music')
plt.ylabel('Level of Reported Insomnia')
```

```python
df.plot.scatter(x='hours', y='ocd', c='orange', alpha = 0.1) 
plt.title('OCD Severity is Not Impacted by Time Spent Listening to Music')
plt.xlabel('Hours/Day Spent Listening to Music')
plt.ylabel('Level of Reported OCD')
```

Just by looking at it, there doesn't seem to be any correlation between time spent listening to music and mental health. The graphs for all four conditions look pretty similar, so it's hard to say if one is impacted more than the other.
By subsetting out our numeric variables, we can also take a look at the correlation coefficients to get a more exact idea.

```python
conditions = df[['hours', 'anxiety', 'depression', 'insomnia', 'ocd']]
```

```python
conditions.corr()
```

The first row of the above table tells us all we need to know. The correlation coefficient is quite low between hours spent listening and any of the conditions. Something interesting, though unrelated, is that anxiety and depression are more closely correlated than any other combination (0.52), albeit the correlation is only moderate.

Now we know that time spent listening has no significant correlation to the severity of mental conditions. But does listening to more music in a day improve your health? Does it worsen it? We can see what the respondents had to say.

Since our dataset is relatively small, I decided to analyze hours against effects as a percentage. If we did the sheer value counts, we could have a situation where there are 100 data points for 2 hours of listening time and 3 data points for 10 hours (or something along those lines). Ideally, we would have enough data for every time slot to get an accurate picture of whether that amount of time helps mental health, but it's difficult to compare the numbers we have unless we do percentages.

```python
timeeffects = df.groupby('hours')['effects'].value_counts() 
timeeffects
```

```python
te = timeeffects.reset_index()
te
```

```python
# Pivot the table to separate 'Yes' and 'No' counts into different columns
te_pivot = te.pivot_table(index=['hours'], columns='effects', values='count', aggfunc='first')

# Flatten the column index
te_pivot.columns = ['Improve', 'No effect', 'Worsen']

# Reset the index to make it a normal DataFrame
te_pivot.reset_index(inplace=True)

tedf = te_pivot.fillna(0)

tedf['imprperc'] = tedf['Improve'] / (tedf['Improve'] + tedf['No effect'] + tedf['Worsen']) * 100
tedf['notimprperc'] = (tedf['No effect'] + tedf['Worsen']) / (tedf['Improve'] + tedf['No effect'] + tedf['Worsen']) * 100

tedf
```

```python
tedf.plot.bar(x = 'hours', y = 'imprperc', color = 'green', legend = False)
plt.title('Music Improves Mental Health Regardless of Listening Time')
plt.ylabel('Improvement Rate (%)')
plt.xlabel('Daily Listening Time (Hours)')
```

Out of curiosity, I also just want to see what the distribution looks like for how many hours the average person listens to per day.

```python
hrs = df['hours'].value_counts()
hrsdf = hrs.reset_index()
sorthrs = hrsdf.sort_values(by = 'hours')
sorthrs
```

```python
colors = ['lightgrey']*5 + ['royalblue']*6 + ['lightgrey']*15
sorthrs.plot.bar(x = 'hours', y = 'count', color = colors, legend = False)
plt.title('Distribution of Listening Time')
plt.ylabel('Frequency')
plt.xlabel('Hours/Day')
```

Most people listen between 1 and 4 hours of music per day.

## Conclusions:
#### The amount of hours that a person spends listening to music has no statistical impact on the severity of their anxiety, depression, insomnia, or OCD. Somebody could listen to music 23 hours of the day or none at all and have the same likelihood of experiencing these conditions.

#### For any amount of time spent listening, music seems to improve mental health across the board, even if it doesn't correlate to the severity of it.

## Question 2: Is somebody more likely to have poor mental health based on their favorite genre?

To answer this question, we'll group the responses by their favorite genre, then find the average rank for each condition and compare between genres.

```python
anx_means = df.groupby('favgenre')['anxiety'].mean() 
sorted_anx_means = anx_means.sort_values()
sorted_anx_means
```

```python
colors = ['green']*0 + ['lightgrey']*14 + ['red']*2
ax = sorted_anx_means.plot.barh(color = colors)
ax.bar_label(ax.containers[0], fmt=' %.1f', label_type = 'edge')
plt.xlim(0,7.5) #Change the limits of the y-axis to better see the difference between genres
plt.title('K-pop and Folk Listeners Report the Most Severe Anxiety')
plt.ylabel('Favorite Genre')
plt.xlabel('Anxiety Severity')
```

```python
dep_means = df.groupby('favgenre')['depression'].mean() 
sorted_dep_means = dep_means.sort_values()
sorted_dep_means
```

```python
colors = ['green']*1 + ['lightgrey']*14 + ['red']*1
ax = sorted_dep_means.plot.barh(color = colors)
ax.bar_label(ax.containers[0], fmt=' %.1f', label_type = 'edge')
plt.xlim(0,7.5)
plt.title('Lofi is more Depressing than Gospel')
plt.ylabel('Favorite Genre')
plt.xlabel('Depression Severity')
```

```python
ins_means = df.groupby('favgenre')['insomnia'].mean() 
sorted_ins_means = ins_means.sort_values()
sorted_ins_means
```

```python
colors = ['green']*0 + ['lightgrey']*14 + ['red']*2
ax = sorted_ins_means.plot.barh(color = colors)
ax.bar_label(ax.containers[0], fmt=' %.1f', label_type = 'edge')
plt.xlim(0,7.5)
plt.title('Lofi and Gospel Listeners Struggle to Sleep')
plt.ylabel('Favorite Genre')
plt.xlabel('Insomnia Severity')
```

```python
ocd_means = df.groupby('favgenre')['ocd'].mean() 
sorted_ocd_means = ocd_means.sort_values()
sorted_ocd_means
```

```python
colors = ['green']*1 + ['lightgrey']*14 + ['red']*1
ax = sorted_ocd_means.plot.barh(color = colors)
ax.bar_label(ax.containers[0], fmt=' %.1f', label_type = 'edge')
plt.xlim(0,7.5)
plt.title("Gospel Lovers Don't Experience Much OCD")
plt.ylabel('Favorite Genre')
plt.xlabel('OCD Severity')
```

At this point, I'm curious why Lofi users tend to report more severe cases of mental conditions. We don't have much information about WHY they listen to Lofi, but we do know whether they listen to music while working. Maybe that can provide us a clue.

```python
working = df.groupby('favgenre')['working'].value_counts()
working
```

```python
workingdf = working.reset_index()
workingdf
```

```python
# Pivot the table to separate 'Yes' and 'No' counts into different columns
wdf_pivot = workingdf.pivot_table(index=['favgenre'], columns='working', values='count', aggfunc='first')

# Flatten the column index
wdf_pivot.columns = ['no', 'yes']

# Reset the index to make it a normal DataFrame
wdf_pivot.reset_index(inplace=True)

workdf = wdf_pivot.fillna(0)

workdf
```

```python
workdf['yesperc'] = workdf['yes'] / (workdf['yes'] + workdf['no']) * 100
workdf['noperc'] = workdf['no'] / (workdf['yes'] + workdf['no']) * 100
```

```python
workdf2 = workdf.sort_values(by = 'yesperc')
workdf2
```

```python
workdf2.plot.barh('favgenre', ['yesperc', 'noperc'], stacked = True, color = ['royalblue', 'lightgrey'])
plt.legend(["Listens While Working", "Doesn't Listen While Working"], loc=(1.05, 0.5))
plt.title("Lofi & Latin are Common to Listen to While Working")
plt.ylabel('Favorite Genre')
plt.xlabel('% of Respondants')
```

All of our respondents who faovir Lofi or Latin listen to music while working. But does that mean listening while working helps to improve mental health? Let's find out by grouping our surveys by who does or does not listen while working, and see how each group said music impacts their mental health.

```python
workeffects = df.groupby('working')['effects'].value_counts()
workeffects
```

```python
wedf = workeffects.reset_index()
wedf
```

```python
# Pivot the table to separate 'Yes' and 'No' counts into different columns
workef_pivot = wedf.pivot_table(index=['working'], columns='effects', values='count', aggfunc='first')

# Flatten the column index
workef_pivot.columns = ['Improve', 'No effect', 'Worsen']

# Reset the index to make it a normal DataFrame
workef_pivot.reset_index(inplace=True)

workef_pivot
```

```python
workef_pivot['improveperc'] = workef_pivot['Improve'] / (workef_pivot['Improve'] + workef_pivot['No effect'] + workef_pivot['Worsen']) * 100
workef_pivot['noneperc'] = workef_pivot['No effect'] / (workef_pivot['Improve'] + workef_pivot['No effect'] + workef_pivot['Worsen']) * 100
workef_pivot['worseperc'] = workef_pivot['Worsen'] / (workef_pivot['Improve'] + workef_pivot['No effect'] + workef_pivot['Worsen']) * 100
workef_pivot
```

```python
workef_pivot.plot.bar(x ='working', y =['improveperc', 'noneperc', 'worseperc'], color = ['green', 'lightgrey', 'red'])
plt.legend(["Improves it", "Has no effect", "Worsens it"], loc=(1.05, 0.5), title = "How does music impact your mental health?")
plt.title("You Should Probably be Listening to Music at Work")
plt.ylabel('% of Respondents')
plt.xlabel('Do you listen to music while working?')
```

## Conclusions: 
#### Those who selected Gospel as their favorite genre reported the lowest average rank in every condition - except for insomnia, where it ranked second highest. Aside from the insomnia, however, those who listen to gospel seem to have better mental health than fans of other genres.
#### Lofi listeners reported the highest levels for every condition but anxiety, where it landed as the fifth highest. It was the farthest ahead of other genres in depression and insomnia. As opposed to gospel, fans of Lofi tend to report worse mental health than other genres.
#### After further analysis, we found that %100 of those who reported Lofi as their favorite genre listen to music while working, and that the vast majority of people in that group say it improves their mental health. This may suggest that Lofi is helpful for reducing mental stress in work environments.

## Question 3: If someone's favorite genre has a higher average BPM, does that correlate to higher levels of anxiety?

We can actually reuse our approach from question 1 to answer this question as well. We just have to map the rank of each condition against BPM instead.

```python
bpmanx = df.groupby('bpm')['anxiety'].mean()
bpmanx
```

```python
badf = bpmanx.reset_index()
badf
```

```python
badf.plot.scatter(x = 'bpm', y = 'anxiety')
```

Without cleaning up the visual at all, we can already see there's a clear outlier stretching the plot. Let's see what it looks like without the outlier.

```python
badf2 = badf[badf['bpm'] < 600]
```

```python
badf2.plot.scatter(x = 'bpm', y = 'anxiety', color = 'red')
plt.title("Higher BPM May Not Cause Anxiety")
plt.ylabel('Reported Anxiety')
plt.xlabel('BPM')
```

Something interesting is that that higher BPM's actually have a wider range of reported anxiety, where as the slower BPM's are more consistently in the same range of anxiety. It doesn't look anything is very correlated here, but let's check the correlation coefficient to confirm.

```python
badf.corr()
```

As we thought, the coefficient is nearly zero.

## Conclusions:
A higher BPM doesn't appear to have any statistical correlation to higher levels of anxiety.

## Question 4: Do different age groups tend to favor certain genres?

To answer this question, we'll group respondents by age and determine the breakdown of their favorite genres in each range.

```python
ages = df[['age', 'favgenre']]
ages
```

I also want to map each genre to a specific color for consistency between each age range.

```python
ages['favgenre'].unique()
```

```python
colormap = {'Video game music' : 'red',
            'Jazz' : 'blue',
            'R&B' : 'purple',
            'K pop' : 'lime',
            'Rock' : 'orange',
            'Country' : 'yellow',
            'EDM' : 'brown',
            'Pop' : 'pink',
            'Hip hop' : 'aqua',
            'Rap' : 'gold',
            'Classical' : 'magenta',
            'Metal' : 'darkgrey',
            'Folk' : 'aquamarine',
            'Lofi' : 'navy',
            'Gospel' : 'yellowgreen',
            'Latin' : 'coral'
}
```

```python
ages['range'] = pd.cut(df['age'], bins=[0, 17, 29, 49, 100], labels=['<18', '18-29', '30-49', '50+'])
```

Before we break it down, let's see what the overall most popular genre is.

```python
agegenres = ages['favgenre'].value_counts().reset_index()

agegenres.plot.pie(x = 'favgenre', y = 'count', labels = agegenres['favgenre'], colors=[colormap[c] for c in agegenres['favgenre']], legend = False, explode = ([0.05] * 16))
plt.title("Favorite Genres")
plt.ylabel("")
```

```python
range1 = ages[ages['range'] == '<18']
range1
```

```python
r1genres = range1['favgenre'].value_counts().reset_index() # Get the counts, then convert them into a dataframe
r1genres
```

```python
r1genres.plot.pie(x = 'favgenre', y = 'count', labels = r1genres['favgenre'], colors=[colormap[c] for c in r1genres['favgenre']], legend = False,explode = ([0.05] * 15))
plt.title("Favorite Genres (Ages <18)")
plt.ylabel("")
```

Pop and rock are the two most popular genres for those under 18. Surprisingly, classical comes in a close 3rd. Let's compare with the other age ranges now.

```python
range2 = ages[ages['range'] == '18-29']
r2genres = range2['favgenre'].value_counts().reset_index()
r2genres.plot.pie(x = 'favgenre', y = 'count', labels = r2genres['favgenre'], colors=[colormap[c] for c in r2genres['favgenre']], legend = False,explode = ([0.05] * 16))
plt.title("Favorite Genres (Ages 18-29)")
plt.ylabel("")
```

Pop and rock come out on top again. Classical falls lower, while metal is more popular.

```python
range3 = ages[ages['range'] == '30-49']
r3genres = range3['favgenre'].value_counts().reset_index()
r3genres.plot.pie(x = 'favgenre', y = 'count', labels = r3genres['favgenre'], colors=[colormap[c] for c in r3genres['favgenre']], legend = False,explode = ([0.05] * 11))
plt.title("Favorite Genres (Ages 30-49)")
plt.ylabel("")
```

People in this age group don't listen to much pop music, mostly rock and metal. Rock seems to be the most popular among all ages so far

```python
range4 = ages[ages['range'] == '50+']
r4genres = range4['favgenre'].value_counts().reset_index()
r4genres.plot.pie(x = 'favgenre', y = 'count', labels = r4genres['favgenre'], colors=[colormap[c] for c in r4genres['favgenre']], legend = False,explode = ([0.05] * 9))
plt.title("Favorite Genres (Ages 50+)")
plt.ylabel("")
```

Once again, rock is the majority winner.

## Conclusions:
Overall, respondents of all ages favor rock, pop, and metal. Those over 50 have the highest preference for gospel music. Those under 18 had the highest preference for classical music, followed by those over 50. Video game music becomes less popular as age increases.

## Question 5: Which mental condition (anxiety, depression, insomnia, OCD) was reported at the highest level? Is music more effective at improving mental health for certain conditions over others?

To answer the first part of our question, we can use the describe() function to view some features of the data.

```python
df.describe()
```

Anxiety has the highest average, followed by depression, then insomnia, then OCD. There were responses of 0 and 10 for all conditions.

To answer the second part of our question, we'll have to do some more manipulations. Lets's assume anybody that reported at least a 5 suffers from a chronic condition. We'll see how many of those people think that music helps them, and see if the numbers differ between conditions.

```python
anxiety = df[df['anxiety'] > 4] # filter out any rows where anxiety is less than 5
```

```python
anxeff = anxiety['effects'].value_counts().reset_index() # convert to a dataframe
anxeff
```

```python
anxeff.plot.pie(x = 'effects', y = 'count', labels = anxeff['effects'], colors = ['green'] + ['lightgrey'] + ['red'], legend = False)
plt.title("Music's Impact on Anxiety")
plt.ylabel("")
```

The vast majority of those with anxiety of 5 or higher say that music improves their mental health. Let's look at the others now.

```python
depression = df[df['depression'] > 4]
depeff = depression['effects'].value_counts().reset_index()

depeff.plot.pie(x = 'effects', y = 'count', labels = depeff['effects'], colors = ['green'] + ['lightgrey'] + ['red'], legend = False)
plt.title("Music's Impact on Depression")
plt.ylabel("")
```

```python
insomnia = df[df['insomnia'] > 4]
inseff = insomnia['effects'].value_counts().reset_index()

inseff.plot.pie(x = 'effects', y = 'count', labels = inseff['effects'], colors = ['green'] + ['lightgrey'] + ['red'], legend = False)
plt.title("Music's Impact on Insomnia")
plt.ylabel("")
```

```python
ocd = df[df['ocd'] > 4]
ocdeff = ocd['effects'].value_counts().reset_index()

ocdeff.plot.pie(x = 'effects', y = 'count', labels = ocdeff['effects'], colors = ['green'] + ['lightgrey'] + ['red'], legend = False)
plt.title("Music's Impact on OCD")
plt.ylabel("")
```

## Conclusions:
All the charts are essentially identical. We can conclude that music is universally beneficial for many people with a variety of mental conditions.
